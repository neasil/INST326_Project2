The MainWindow class is a subclass of Tkinter’s tk.Tk class and thus inherits its attributes and methods, while allowing us to customize the new window objects to our needs. These new window objects will represent “notebooks” or collections of notes, and allow us to work with those notes. (I have named it MainWindow because this class definition could be used to create any kind of main window in Tkinter. We are using it to represent notebooks in this application, but you might use it for other purposes in onther applications.)


The TopWindow class creates a new top window in Tkinter, which is essentially a form for entering the title, text, links, and tags for our note. When we submit the note, this “form” object has a method that creates the note’s metadata and a new Note object. That note object is appended to the list of all notes, which is an attribute of the notebook (MainWindow) class.
The Note class creates note objects that contain the  title, text, links, tags, and metadata for each note.

The code cell below contains the imports you will need; the top level structure for the three classes to get you started; and the execution code at the bottom. 

```python
# Imports
import tkinter as tk
from tkinter import ttk
from tkinter import filedialog
import datetime # Module for working with dates and times
from functools import partial # Module for passing arguments in display_notebook command

# The MainWindow class creates a custom GUI window based on the tkinter window (tk.Tk)
# It has an __init__() method, and three additional methods (new_note(), open_notebook(), and save_notebook())
# These methods correspond to new, open, and save buttons in the window.
# The new_note method calls the NoteForm class to create a new note form top level window.

class MainWindow(tk.Tk):
    def __init__(self, notebook = None, notes = None): # Initialize the main window
        super().__init__() # Initialize as a tkinter window
        
        self.geometry("600x400") # Set the default window size
        self.title('Notebook') # Set the default window title
        self.notebook = [] # Initialize an attribute named 'notebook' as an empty list
        self.notes = [] # Initialize 'list' attribute as an empty list      
        
        
    def new_note(self): # For creating new notes
        
        # Initialize instance of NoteForm
        global note_window
        note_window = NoteForm(self, self.notebook, self.notes)
        
        # Title label
        title_label = tk.Label(note_window, bg = 'light gray', text = 'Note Title:')
        title_label.grid(padx = 10, pady = 10, row = 1, column = 0, sticky = 'e')
        
        # Text label
        text_label = tk.Label(note_window, bg='light gray', text='Note Text:')
        text_label.grid(padx = 10, pady = 10, row = 2, column = 0, sticky = 'e')
        
        # Link label
        link_label = tk.Label(note_window, bg = 'light gray', text = 'Note Link:')
        link_label.grid(padx = 10, pady = 10, row = 3, column = 0, sticky = 'e')
        
        # Tag label
        tag_label = tk.Label(note_window, bg = 'light gray', text = 'Note Tags:')
        tag_label.grid(padx = 10, pady = 10, row = 4, column = 0, sticky = 'e')

        # Create note title entry field
        noteform_title = note_window.form_title
        noteform_title.grid(padx = 10, pady = 10, row = 1, column = 1, sticky = 'w')
        noteform_title.insert(0, 'New note title') # Adds default title
    
        # Create note text field
        noteform_text = note_window.text
        noteform_text.grid(padx = 10, pady = 10, row = 2, column = 1)
        noteform_text.insert('1.0', "Enter text here") # Adds default text
    
        # Create link text field
        noteform_link = note_window.link
        noteform_link.grid(padx = 10, pady = 10, row = 3, column = 1)
    
        # Create tags text field
        noteform_tags = note_window.tags
        noteform_tags.grid(padx = 10, pady = 10, row = 4, column = 1) 

        # Create 'submit' button
        b1 = tk.Button(note_window, text = 'submit', command = note_window.submit)
        b1.grid(padx = 10, pady = 10, row = 6, column = 1, sticky = 'w')
        
        return None

    
    def open_notebook(self): # For opening previously saved notes in new window 
        
        # Open file path
        filepath = filedialog.askopenfilename(initialdir = "/Users/baihanliang/INST326-0104",
                                         filetypes=[("text files", "*.txt"), 
                                                    ("csv files", ".csv"),
                                         ("all files", "*.*")])
        file = open(filepath, "r")
        file_list = file.read()
        
        # Open new window for file text to be displayed
        notebook_window = tk.Toplevel() 
        notebook_window.geometry("400x300")
        notebook_window.title("Note")
        display_text = file_list
        text_label = tk.Label(notebook_window, text = display_text)
        text_label.grid(padx = 10, pady = 10, row = 2, column = 5)
        file.close()
        
        return None
   
    def display_notebook(self, note_to_display): # For displaying created notes in new window
        
        # Create new window
        notebook_window = tk.Toplevel() 
        notebook_window.geometry("400x300")
        notebook_window.title("Note")
        display_text = f'\n{note_to_display.note_title}\n{note_to_display.note_text}\n link: {note_to_display.note_link}\n tags: {note_to_display.note_tags}\n{note_to_display.note_meta}\n'
        text_label = tk.Label(notebook_window, text = display_text)
        text_label.grid(padx = 10, pady = 10, row = 2, column = 3)

    def save_notebook(self): # For saving notes to computer
        
        # Ask to save note as file
        file = filedialog.asksaveasfile(initialdir = "/Users/baihanliang/INST326-0104",
                                          defaultextension = ".txt", 
                                          filetypes=[("text file", ".txt"),
                                         ("all files", ".*")])
        
        # Iterate through each note in 'notes', writing each one to the file
        for note in note_window.notes:
            file.write(f'\n{note.note_title}\n{note.note_text}\n link: {note.note_link}\n tags: {note.note_tags}\n{note.note_meta}\n')
        file.close()
        
        return None

# the NoteForm() class creates a Toplevel window that is a note form containing fields for
# data entry for title, text, link, and tags. It also calculates a meta field with date, time, and timezone
# the Noteform class has an __init__() method, and a submit() method that is called by a submit button
# the class may contain additional methods to perform tasks like calculating the metadata, for example
# the submit method calls the MakeNote class that transforms the entered data into a new note object.

class NoteForm(tk.Toplevel):
    
    def __init__(self, master = None, notebook = None, notes = None, form_title = None, text = None, link = None, tags = None, date_and_time = None, timezone = None, meta = None): # initialize the new object
        super().__init__(master) # Initialize it as a Toplevel window
        
        self.geometry("600x350")
        self.title("Note Form")
        self.config(bg = "light gray")
        self.notebook = notebook
        self.notes = notes
        self.form_title = tk.Entry(self, width = 80)
        self.text = tk.Text(self, height = 10, width = 60)
        self.link = tk.Entry(self, width = 80)
        self.tags = tk.Entry(self, width = 80)
        self.date_and_time = datetime.datetime.now()
        self.timezone = self.date_and_time.astimezone().tzinfo
        self.meta = f'created {self.date_and_time} {self.timezone}'
    
        
    def submit(self): # For storing created notes in 'notes'
        
        # Creating note_dict, which contains user inputs for note attributes
        title = self.form_title.get()
        text = self.text.get('1.0', 'end').strip('\n')
        link = self.link.get()
        tags = self.tags.get()
        note_dict = {'title':title, 'text':text, 'link': link, 'tags': tags, 'meta': self.meta}
        
        # Create note object and append to 'notes' and 'notebook'
        new_note = MakeNote(note_dict)
        self.notes.append(new_note)
        
        # Print note contents
        printable = f'{title}\n{text}\n{link}\n{tags}\n{self.meta}'
        self.notebook.append(printable)
        
        # Add button representing note to main window
        filetext = f"{title}\n{text}\n link: {link}\n tags: {tags}\n{self.meta}"
        note_button = tk.Button(main_window, text = title, command = partial(main_window.display_notebook, new_note))
        note_button.grid(padx = 10, pady= 10, row = 2, column = note_window.notes.index(new_note)+1)
        
        return printable
        

    
# The MakeNote class takes a dictionary containing the data entered into the form window,
# and transforms it into a new note object.

class MakeNote():
    def __init__(self, note_dict = None, note_title = None, note_text = None, note_link = None, note_tags = None, note_meta = None):
        
        self.note_title = note_dict.get("title")
        self.note_text = note_dict.get("text")
        self.note_link = note_dict.get("link")
        self.note_tags = note_dict.get("tags")
        self.note_meta = note_dict.get("meta")


# Main execution
if __name__ == '__main__':
    
    # Create a notebook
    main_window = MainWindow() 
    
    # Label for placing buttons
    button_label = tk.Label(main_window)
    button_label.grid(padx = 10, pady = 10, row = 8, column = 3)
    
    # New note button
    new_note = tk.Button(button_label, text = 'new', command = main_window.new_note)
    new_note.grid(padx=10, pady=10, row=5, column=0)
    
    # Open note button
    open_notebook = tk.Button(button_label, text='open', command = main_window.open_notebook)
    open_notebook.grid(padx=10, pady=10, row=5, column=1)
    
    # Save note button
    save_notebook = tk.Button(button_label, text='save', command = main_window.save_notebook)
    save_notebook.grid(padx=10, pady=10, row=5, column=2) 
    
    main_window.mainloop()
```
